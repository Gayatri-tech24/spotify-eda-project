Spotify songs data analysis

## Insights

- Most songs in the dataset have low popularity scores.
- Only a small percentage of songs achieve extremely high popularity.
- Danceability does not show a strong correlation with popularity.
- Energy and loudness are positively correlated.
- Popularity appears to depend on multiple musical and external factors.

```python
#loading the dastaset
import pandas as pd
df=pd.read_csv('spotify-tracks.csv')
```

```python
#viewing the first 5 rows of the dataset
df.head()
```

```python
df.info()
```

```python
df.describe()
```

```python
# to find the number of null values in each column
df.isnull().sum()
```

```python
df=df.dropna()
```

```python
print(df["artists"].value_counts().head(10))
```

```python
top10=df.sort_values(by="popularity",ascending=False).drop_duplicates(subset="track_name")
```

```python
top=top10[["track_name","artists","album_name","popularity"]].head(10)
print(top)
```

```python
#popularity distribution graph
import matplotlib.pyplot as plt
import seaborn as sns
plt.figure(figsize=(10,8))
sns.histplot(data=df["popularity"],bins=50,kde=True)
plt.xlabel("Popularity")
plt.ylabel("no of songs")
plt.title("Popularity Distribution")
plt.show()
plt.savefig("images/popularity_distribution.png")
```

The popularity distribution is heavily skewed toward lower popularity values, indicating that most songs in the dataset are not highly popular. Only a smaller percentage of tracks achieve very high popularity scores.

```python
#correlation graphs 
plt.figure(figsize=(10,8))
sns.scatterplot(x="popularity", y="danceability",data=df,hue="danceability",alpha=0.5)
plt.title("Danceability vs Popularity")
plt.show()
plt.savefig("images/scatterplot.png")
```

The scatterplot shows no strong linear relationship between danceability and popularity. Songs with varying danceability levels can achieve high popularity, suggesting that popularity depends on multiple factors beyond danceability alone.
